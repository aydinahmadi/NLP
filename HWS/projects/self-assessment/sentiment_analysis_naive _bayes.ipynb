```python
import re
#removing the useless characters
def clean_text(text):
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    cleaned_text = re.sub(r'\d+', '', cleaned_text)
    cleaned_text = cleaned_text.lower()
    return cleaned_text
```

```python
import os
def load_txt_files2(directory):
    word_counts = {}
    file_names = os.listdir(directory)
    counter = 0
    for filename in file_names:
        counter += 1
        print(counter)
        file_path = os.path.join(directory, filename)
        file = open(file_path, "r",  encoding="utf8")
        file_content = file.read()
        file_content = clean_text(file_content)
        file_content = file_content.split()
        # Loop through each word and update its count in the dictionary
        for word in file_content:
            if word in word_counts:
                word_counts[word] += 1
            else:
                word_counts[word] = 1
    return word_counts, counter

#loading
pos_vocab_dict, splited_train_pos_words = load_txt_files2('D:\\Koc\\NLP Sahin\\dataset\\train\\pos')

neg_vocab_dict , splited_train_neg_words = load_txt_files2('D:\\Koc\\NLP Sahin\\dataset\\train\\neg')
```

```python
len(pos_vocab_dict), len(neg_vocab_dict), splited_train_pos_words, splited_train_neg_words
```

```python
#prior estimation
q_hat_positive = splited_train_pos_words/ (splited_train_pos_words + splited_train_neg_words)

q_hat_negative = splited_train_neg_words/ (splited_train_pos_words + splited_train_neg_words)
q_hat_positive, q_hat_negative
```

```python
num_pos_vocab = len(pos_vocab_dict)
num_neg_vocab = len(neg_vocab_dict)
num_vocab = num_pos_vocab + num_neg_vocab
num_pos_vocab, num_neg_vocab, num_vocab
```

```python
#count words occurrences in a dictionary
def count_dict(dictionary):
    counter = 0
    for word in dictionary:
        counter += dictionary[word]
    
    return counter

pos_count_dict = count_dict(pos_vocab_dict)
neg_count_dict = count_dict(neg_vocab_dict)
pos_count_dict, neg_count_dict
```

```python
#load test data
import os
def load_txt_files_test(directory):
    test_list = []
    file_names = os.listdir(directory)
    counter = 0
    for filename in file_names:
        word_counts = {}
        counter += 1
        print(counter)
        file_path = os.path.join(directory, filename)
        file = open(file_path, "r",  encoding="utf8")
        file_content = file.read()
        file_content = clean_text(file_content)
        file_content = file_content.split()
        # Loop through each word and update its count in the dictionary
        for word in file_content:
            if word in word_counts:
                word_counts[word] += 1
            else:
                word_counts[word] = 1
        test_list.append(word_counts)
        
    return test_list

#loading
test_pos_list = load_txt_files_test('D:\\Koc\\NLP Sahin\\dataset\\test\\pos')

test_neg_list  = load_txt_files_test('D:\\Koc\\NLP Sahin\\dataset\\test\\neg')
```

```python
count_total_vocab = len(pos_vocab_dict) + len(neg_vocab_dict)
count_total_vocab
```

```python
import numpy as np
#laplace smoothing
#probability estimation given classes and words
def probability_estimation(test_list_dict, dictionary, prior, count_total_vocab, count_in_dict):
    probabilities = []
    for test_dict in test_list_dict:
        probability_of_document = 0
        probability_of_words = 0
        for test_word in test_dict:
    #if word exists in dictionary calculate the probability by its count else laplace smootihng
    #log(count(wordi, category) + 1) - log(count(category) + |v|)
            prob_word = 1
            if test_word in dictionary:
                prob_word += dictionary[test_word]
            else:
                prob_word = 1
            probability_of_words += np.log(prob_word) - np.log((count_in_dict + count_total_vocab))
            
        #log(p(category|document)) = log(p(category)) + sigma log(p(wordi) | document)
        probability_of_document += np.log(prior)
        probability_of_document += probability_of_words
        probabilities.append(probability_of_document)
                

    return probabilities
```

```python
pos_pos_prob = probability_estimation(test_pos_list, pos_vocab_dict, q_hat_positive, num_vocab, pos_count_dict)
pos_neg_prob = probability_estimation(test_pos_list, neg_vocab_dict, q_hat_negative, num_vocab, neg_count_dict)

counter_true_positives = 0
for i in range(12500):
    if pos_neg_prob[i] < pos_pos_prob[i]:
        counter_true_positives += 1

print("Number of true positives: {}".format(counter_true_positives))
```

```python
neg_pos_prob = probability_estimation(test_neg_list, pos_vocab_dict, q_hat_positive, num_vocab, pos_count_dict)
neg_neg_prob = probability_estimation(test_neg_list, neg_vocab_dict, q_hat_negative, num_vocab, neg_count_dict)
counter_true_negatives = 0

for i in range(12500):
    if neg_pos_prob[i] < neg_neg_prob[i]:
        counter_true_negatives += 1

print("Number of true negatives:{}".format(counter_true_negatives))
```

```python
print("accuracy: {}%".format(float((counter_true_negatives+counter_true_positives)/ 25000) * 100))
```
